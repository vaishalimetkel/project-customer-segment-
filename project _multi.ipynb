```python
#clustering process 
# between age and spending score(1-100)

x1 = df.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans 
wcss =[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init ="k-means++")
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker ="o")
plt.xlabel("k value ")
plt.ylabel("wcss")
plt.show()
```

```python
# no of cluster we have to create 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
x1 = df.loc[:, ["Age", "Spending Score (1-100)"]].values

wcss = []  # Initialize the list to store the Within-Cluster Sum of Squares

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="o")  # Changed marker to 'o'
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.title("The Elbow Method")
plt.show()
  




```

```python
# data divide into 4 clusters  0 to 3
kmeans = KMeans(n_clusters=4)

label = kmeans.fit_predict(x1)

print(label)
```

```python
# finding centroids
print(kmeans.cluster_centers_)
```

```python
# visualization of clusters 
plt.scatter(x1[:,0],x1[:,1],c = kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], color ='black')
plt.title('cluster of customers')
plt.xlabel('Age')
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame named 'df' with the data

# Drop any non-numeric columns or columns that are not suitable for clustering
X3 = df.select_dtypes(include='number')

# Standardize the data to have zero mean and unit variance (important for k-means)
scaler = StandardScaler()
X3_scaled = scaler.fit_transform(X3)

# Initialize an empty list to store the WCSS for each value of k
wcss = []

# Loop through different values of k from 1 to 10
for k in range(1, 11):
    # Create a KMeans clustering model with 'k' clusters using the 'k-means++' initialization method
    kmeans = KMeans(n_clusters=k, init='k-means++')
    
    # Fit the model to the standardized data 'X3_scaled'
    kmeans.fit(X3_scaled)
    
    # Append the within-cluster sum of squares (WCSS) to the list
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against different values of k
plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="o")  # Changed marker to 'o'
plt.xlabel("k value")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Assuming you have already defined and fitted the KMeans model as "kmeans" and have the DataFrame "df" containing the data

# Create a new column 'label' in the DataFrame to store the cluster labels
clusters = kmeans.fit_predict(df[["Age", "Annual Income ($)", "Spending Score (1-100)"]])
df["label"] = clusters

# Initialize the 3D figure
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection="3d")

# Create a list of colors for different clusters
colors = ['blue', 'red', 'green', 'orange', 'purple']

# Plot each cluster with a different color
for label, color in enumerate(colors):
    ax.scatter(df["Age"][df["label"] == label], 
               df["Annual Income ($)"][df["label"] == label], 
               df["Spending Score (1-100)"][df["label"] == label], 
               c=color, s=60, label=f"Cluster {label}")

# Set labels and title
ax.set_xlabel("Age")
ax.set_ylabel("Annual Income ($)")
ax.set_zlabel("Spending Score (1-100)")
ax.set_title("Customer Segmentation")

# Add a legend
ax.legend()

# Set the viewpoint for a better angle (optional)
ax.view_init(30, 185)

# Show the plot
plt.show()
```
